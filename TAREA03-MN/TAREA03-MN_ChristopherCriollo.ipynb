**1.** Utilice **aritm√©tica de corte de tres d√≠gitos** para calcular las siguientes sumas. Para cada parte, **¬øqu√© m√©todo es m√°s preciso y por qu√©?**

**a.** $$\sum_{i=1}^{10} \left( \frac{1}{i^2} \right) \text{ primero por } \frac{1}{1} + \frac{1}{4} + \cdots + \frac{1}{100} \text{ y luego por } \frac{1}{100} + \frac{1}{81} + \cdots + \frac{1}{1}$$

```python
def three_digit_chop(x):
    if x == 0:
        return 0
    
    # Para n√∫meros muy peque√±os
    if abs(x) < 0.001:
        s = f"{abs(x):.10f}"
        for i, char in enumerate(s):
            if char not in ['0', '.']:
                significant = s[i:i+3]
                result = float(significant) * (10 ** -(i-2))
                return result if x > 0 else -result
    
    # Para n√∫meros normales
    s = f"{abs(x):.10f}"
    if '.' in s:
        integer_part, decimal_part = s.split('.')
        if integer_part != '0':
            if len(integer_part) >= 3:
                result = float(integer_part[:3])
            else:
                digits_needed = 3 - len(integer_part)
                result = float(integer_part + decimal_part[:digits_needed]) / (10 ** digits_needed)
        else:
            count_zeros = 0
            for char in decimal_part:
                if char == '0':
                    count_zeros += 1
                else:
                    break
            significant = decimal_part[count_zeros:count_zeros+3]
            result = float(significant) * (10 ** -(count_zeros + 3))
    else:
        result = float(s[:3]) if len(s) > 3 else float(s)
    
    return result if x >= 0 else -result

def suma_con_corte(terminos):
    suma = 0
    for termino in terminos:
        termino_chop = three_digit_chop(termino)
        suma_chop = three_digit_chop(suma + termino_chop)
        suma = suma_chop
    return suma

print("LITERAL a: ‚àë(1/i¬≤) desde i=1 hasta 10")

terminos_asc_a = [1/(i**2) for i in range(1, 11)]
terminos_desc_a = [1/(i**2) for i in range(10, 0, -1)]

# Valores exactos
suma_exacta_a = sum(terminos_asc_a)

# Sumas con corte
suma_asc_a = suma_con_corte(terminos_asc_a)
suma_desc_a = suma_con_corte(terminos_desc_a)

print(f"Suma exacta: {suma_exacta_a:.10f}")
print(f"Orden ascendente: {suma_asc_a:.6f}")
print(f"Orden descendente: {suma_desc_a:.6f}")

# Errores
error_asc_a = abs(suma_exacta_a - suma_asc_a)
error_desc_a = abs(suma_exacta_a - suma_desc_a)

print(f"Error orden ascendente: {error_asc_a:.6e}")
print(f"Error orden descendente: {error_desc_a:.6e}")

if error_asc_a < error_desc_a:
    print("\nM√ÅS PRECISO: Orden ascendente")
    print("  RAZ√ìN: Al sumar de menor a mayor, se minimiza la p√©rdida de")
    print("         d√≠gitos significativos cuando se suman n√∫meros peque√±os")
    print("         a acumulados grandes.")
else:
    print("\nM√ÅS PRECISO: Orden descendente")
    print("  RAZ√ìN: Al sumar de mayor a menor, los t√©rminos m√°s grandes")
    print("         se suman primero, reduciendo errores de redondeo.")

# Demostraci√≥n de los t√©rminos con corte
print("\nT√©rminos individuales con corte de 3 d√≠gitos:")
print("Orden ascendente:")
for i, term in enumerate(terminos_asc_a, 1):
    print(f"  1/{i}¬≤ = {term:.6f} -> {three_digit_chop(term):.6f}")
```

**2.** La serie de Maclaurin para la funci√≥n arcotangente converge para $-1 \le x \le 1$ y est√° dada por
    $$\arctan x = \lim_{n\to\infty} P_n(x) = \lim_{n\to\infty} \sum_{i=1}^{n} (-1)^{i+1} \frac{x^{2i-1}}{2i-1}$$

    

**a.** Utilice el hecho de que $\tan \pi/4 = 1$ para determinar el n√∫mero $n$ de t√©rminos de la serie que se necesita sumar para garantizar que $$|4P_n(1) - \pi| < 10^{-3}$$

```python
import math

def arctan_maclaurin(x, n_terms):
    result = 0.0
    for i in range(1, n_terms + 1):
        term = ((-1) ** (i + 1)) * (x ** (2 * i - 1)) / (2 * i - 1)
        result += term
    return result

def encontrar_terminos_para_precision_10_3():
    print("LITERAL 2a: T√©rminos para |4P_n(1) - œÄ| < 10^{-3}")
    
    pi_real = math.pi
    objetivo_error = 1e-3
    n = 1
    
    print(f"œÄ real: {pi_real:.10f}")
    print(f"Error objetivo: {objetivo_error}")
    print("\nBuscando n√∫mero de t√©rminos necesarios...")
    print("-" * 60)
    
    while True:
        # Calcular arctan(1) con n t√©rminos
        arctan_approx = arctan_maclaurin(1, n)
        # Calcular œÄ aproximado: 4 * arctan(1)
        pi_approx = 4 * arctan_approx
        # Calcular error
        error = abs(pi_approx - pi_real)
        
        print(f"n = {n:3d}: œÄ_approx = {pi_approx:.8f}, error = {error:.2e}")
        
        # Verificar si cumple con el criterio de error
        if error < objetivo_error:
            print("-" * 60)
            print(f"‚úì RESULTADO: Se necesitan {n} t√©rminos")
            print(f"  Error final: {error:.2e} < {objetivo_error}")
            print(f"  œÄ aproximado: {pi_approx:.8f}")
            print(f"  œÄ real:      {pi_real:.8f}")
            break
        
        n += 1
        
        # Prevenir bucle infinito
        if n > 1000:
            print("Advertencia: No se alcanz√≥ la precisi√≥n en 1000 t√©rminos")
            break
    
    return n

# Ejecutar el an√°lisis
n_terminos_10_3 = encontrar_terminos_para_precision_10_3()

# Demostraci√≥n detallada de los primeros t√©rminos
print("\n" + "=" * 60)
print("DEMOSTRACI√ìN DETALLADA DE LOS PRIMEROS T√âRMINOS")
print("=" * 60)

print("n\tT√©rmino\t\tSuma Parcial\t4*Suma\t\tError")
print("-" * 80)

suma_parcial = 0
for i in range(1, min(11, n_terminos_10_3 + 3)):
    termino = ((-1) ** (i + 1)) * (1 ** (2 * i - 1)) / (2 * i - 1)
    suma_parcial += termino
    pi_aprox = 4 * suma_parcial
    error = abs(pi_aprox - math.pi)
    
    print(f"{i}\t{termino:+.6f}\t{suma_parcial:.6f}\t{pi_aprox:.6f}\t{error:.6f}")
```

**b.** El lenguaje de programaci√≥n C++ requiere que el valor de $\pi$ se encuentre dentro de $10^{-10}$. **¬øCu√°ntos t√©rminos de la serie se necesitar√≠an sumar para obtener este grado de precisi√≥n?**

```python
import math

def arctan_maclaurin_optimizado(x, n_terms):
    result = 0.0
    x_power = x  # x^(2i-1)
    
    for i in range(1, n_terms + 1):
        term = ((-1) ** (i + 1)) * x_power / (2 * i - 1)
        result += term
        x_power *= x * x  # Actualizar potencia: x^(2i+1) = x^(2i-1) * x^2
    return result

def encontrar_terminos_para_precision_10_10():
    print("LITERAL 2b: T√©rminos para |4P_n(1) - œÄ| < 10^{-10}")
    print("=" * 60)
    
    pi_real = math.pi
    objetivo_error = 1e-10
    n = 1
    
    print(f"œÄ real: {pi_real:.15f}")
    print(f"Error objetivo: {objetivo_error:.1e}")
    print("\nBuscando n√∫mero de t√©rminos necesarios...")
    print("-" * 80)
    
    # Solo mostrar algunos puntos para no saturar
    puntos_mostrar = [1, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
    
    while True:
        # Calcular aproximaci√≥n
        arctan_approx = arctan_maclaurin_optimizado(1, n)
        pi_approx = 4 * arctan_approx
        error = abs(pi_approx - pi_real)
        
        # Mostrar progreso en puntos espec√≠ficos
        if n in puntos_mostrar or error < objetivo_error:
            print(f"n = {n:5d}: œÄ_approx = {pi_approx:.12f}, error = {error:.2e}")
        
        # Verificar criterio
        if error < objetivo_error:
            print("-" * 80)
            print(f"‚úì RESULTADO: Se necesitan {n} t√©rminos")
            print(f"  Error final: {error:.2e} < {objetivo_error:.1e}")
            print(f"  œÄ aproximado: {pi_approx:.12f}")
            print(f"  œÄ real:      {pi_real:.12f}")
            
            # An√°lisis adicional
            print(f"\n  √öltimo t√©rmino: {1/(2*n-1):.2e}")
            print(f"  El t√©rmino n={n} es aproximadamente {1/(2*n-1):.2e}")
            break
        
        n += 1
        
        # L√≠mite de seguridad
        if n > 1000000:
            print("Advertencia: No se alcanz√≥ la precisi√≥n en 1,000,000 t√©rminos")
            break
    
    return n

# Ejecutar el an√°lisis
n_terminos_10_10 = encontrar_terminos_para_precision_10_10()

# An√°lisis te√≥rico del error
print("\n" + "=" * 60)
print("AN√ÅLISIS TE√ìRICO DEL ERROR")
print("=" * 60)

print("Para la serie alternante arctan(1) = 1 - 1/3 + 1/5 - 1/7 + ...")
print("El error despu√©s de n t√©rminos est√° acotado por el primer t√©rmino omitido:")
print("|Error| ‚â§ 1/(2n+1)")

# Calcular n te√≥rico para error < 1e-10
n_teorico = (1 / (2 * 1e-10) - 1) / 2
print(f"\nPara error < 1e-10, necesitamos 1/(2n+1) < 1e-10")
print(f"Esto implica: 2n + 1 > 1e10")
print(f"n > (1e10 - 1)/2 ‚âà {n_teorico:.0f}")

print(f"\nComparaci√≥n:")
print(f"  n encontrado: {n_terminos_10_10}")
print(f"  n te√≥rico: ~{n_teorico:.0f}")
```

```python
def raices(a : float, b : float, c : float) -> tuple[float, float] | float | tuple[complex, complex]:
    discriminante = b**2 - 4*a*c
    
    if (discriminante == 0):
        raiz = (-b + (discriminante)**0.5) / 2*a
        return raiz
    
    elif (discriminante > 0):
        raiz1 = (-b + (discriminante)**0.5) / 2*a
        raiz2 = (-b - (discriminante)**0.5) / 2*a
        return raiz1, raiz2
    
    else:
        raiz1 = complex(-b / 2*a, (abs(discriminante))**0.5 * 100 // 2*a / 100)
        raiz2 = complex(-b / 2*a, -(abs(discriminante))**0.5 * 100 // 2*a / 100)
        return raiz1, raiz2


resultado = raices(1.5, -2, 1)
print('La/s raiz/ices de la ecuacion cuadratica es/son: ', resultado)
    
```
